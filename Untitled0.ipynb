# Sine wave superposition

We're looking at some discrete positions of a massless rope of 10 m length. 

This wave has the periodicity of 2 periods per 10 m and an amplitude of 1

```
import plotly.graph_objects as go
import numpy as np
x = np.linspace(0,10,100)
frames=[go.Frame(data=[go.Scatter(x=x, y=np.sin(1*np.pi*x/10)*np.sin(t), mode='markers')]) for t in np.linspace(0, 2*np.pi, 100)]
fig = go.Figure(
    data=[go.Scatter(x=[0, 10], y=[0, 0], mode='markers')],
    layout=go.Layout(
        xaxis=dict(range=[0, 10], autorange=False),
        yaxis=dict(range=[-1, 1], autorange=False),
        title="Start Title",
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label="Play",
                          method="animate",
                          args = [None, {"frame": {"duration": 50, 
                                                    "redraw": True},
                                                    "fromcurrent": True, 
                                                    "transition": {"duration": 0}}],)])]
    ),
    frames = frames
)
fig.show()
```

Now let's look at another sine wave, with the periodicity of 4 periods per 10 m and an amplitude of 0.5

```
x = np.linspace(0,10,100)

frames=[go.Frame(data=[go.Scatter(x=x, y=0.5*np.sin(3*np.pi*x/10)*np.sin(t), mode='markers')]) for t in np.linspace(0, 2*np.pi, 100)]
fig = go.Figure(
    data=[go.Scatter(x=[0, 10], y=[0, 0], mode='markers')],
    layout=go.Layout(
        xaxis=dict(range=[0, 10], autorange=False),
        yaxis=dict(range=[-1, 1], autorange=False),
        title="Start Title",
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label="Play",
                          method="animate",
                          args = [None, {"frame": {"duration": 50, 
                                                    "redraw": True},
                                                    "fromcurrent": True, 
                                                    "transition": {"duration": 0}}],)])]
    ),
    frames = frames
)
fig.show()
```

And one more, with an amplitude of 0.25 and a periodicity of 8 periods per 10 m.

```

x = np.linspace(0,10,100)
frames=[go.Frame(data=[go.Scatter(x=x, y=np.sin(5*np.pi*x/10)*np.sin(t), mode='markers')]) for t in np.linspace(0, 2*np.pi, 100)]

fig = go.Figure(
    data=[go.Scatter(x=[0, 10], y=[0, 0], mode='markers')],
    layout=go.Layout(
        xaxis=dict(range=[0, 10], autorange=False),
        yaxis=dict(range=[-2, 2], autorange=False),
        title="Start Title",
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label="Play",
                          method="animate",
                          args = [None, {"frame": {"duration": 50, 
                                                    "redraw": True},
                                                    "fromcurrent": True, 
                                                    "transition": {"duration": 0}}],)])]
    ),
    frames = frames
)
fig.show()
```

Now let's add them all together

```

x = np.linspace(0,10,100)
frames=[go.Frame(data=[go.Scatter(x=x, y=np.sin(1*np.pi*x/10)*np.sin(t) - np.sin(3*np.pi*x/10)*np.sin(3*t) + np.sin(5*np.pi*x/10)*np.sin(5*t), mode='markers')]) for t in np.linspace(0, 2*np.pi, 100)]

fig = go.Figure(
    data=[go.Scatter(x=[0, 10], y=[0, 0], mode='markers')],
    layout=go.Layout(
        xaxis=dict(range=[0, 10], autorange=False),
        yaxis=dict(range=[-2, 2], autorange=False),
        title="Start Title",
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label="Play",
                          method="animate",
                          args = [None, {"frame": {"duration": 50, 
                                                    "redraw": True},
                                                    "fromcurrent": True, 
                                                    "transition": {"duration": 0}}],)])]
    ),
    frames = frames
)
fig.show()
```
